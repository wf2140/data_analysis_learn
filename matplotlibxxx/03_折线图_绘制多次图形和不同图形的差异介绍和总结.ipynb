```python
from matplotlib import pyplot as plt
import matplotlib
#windws和linux设置字体的放
font ={'family':'MicroSoft YaHei','weight':'bold'}
matplotlib.rc("font",**font)
matplotlib.rc("font",family='MicroSoft YaHei',weight="bold")
```

```python
x = range(11,31)
y1 = [1,0,1,1,2,4,3,2,3,4,4,5,6,5,4,3,3,1,1,1]
y2 = [2,3,4,5,4,3,5,6,8,2,2,3,4,6,2,1,3,0,2,1] 
y3 = [1,3,2,4,2,4,3,1,1,1,1,3,4,2,1,1,2,3,1,4]
```

- 在绘制的时候指定即可color='ri，线条颜色
- linestyle='--'线条风格
- 线条粗细linewidth=5，
- alpha=0.5透明度

```python
plt.figure(figsize=(20,8),dpi=80)

_x = ['%s岁'%(i)for i in x]
plt.xticks(list(x),_x,rotation=-45)
plt.title("每年朋友数量")
plt.xlabel("年纪")
plt.ylabel('对象数量')
plt.grid(linestyle=':')
plt.plot(x,y1,label='我',color = 'cyan',linestyle=':')
plt.plot(x,y2,label='同桌',linestyle='-.')
plt.plot(x,y3,label='同',linestyle='--')
# loc
plt.legend()
```

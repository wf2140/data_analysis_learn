### 为什么无法显示中文:matplotlib默认不支持中文字符，因为默认的英文字体无法显示汉字
- 查看linux/mac下面支持的字体:fc-list →查看支持的字体
  - fc-list :lang=zh →查看支持的中文(冒号前面有空格)
- 那么问题来了:如何修改matplotlib的默认字体?
  - 通过matplotlib.rc可以修改,具体方法参见源码(windows/linux)
  - 通过matplotlib 下的font_manager可以解决(windows/linux/mac)

```python
from matplotlib import pyplot as plt
import random
import matplotlib
#windws和linux设置字体的放
font ={'family':'MicroSoft YaHei','weight':'bold'}
matplotlib.rc("font",**font)
matplotlib.rc("font",family='MicroSoft YaHei',weight="bold")
# matplotlib.rcParams['font.family'] = 'SimHei'
# matplotlib.rcParams['font.size'] = 10
```

```python
x = range(0,120)
y = [random.randint(20,35)for i in range(120)]
```

```python
plt.figure(figsize=(20,8),dpi=80)
_tick_lable = ['%s时%s分'%(i//60 +10,i%60) for i in range(0,120,3)]
plt.xticks(list(x)[::3],_tick_lable,rotation=-45)
plt.xlabel('时间')
plt.ylabel("温度:单位℃")
plt.title("十点到十二点每分钟温度变化情况")
plt.plot(x,y)
```

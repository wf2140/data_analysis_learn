```python
from matplotlib import pyplot as plt
```

```python
#数据在x和y轴的位置,是一个可迭代对象
x = range(2,26,2)
y = [15,13,14.5,17,20,25,26,26,27,22,18,15]
```

 - x轴和y轴的数据一起组成了所有要绘制出的坐标 
 - 分别是(2,15),(4,13),(6,14.5),(8,17).t(x,y)
 - 传入x和y,通过plot绘制出折线图在执行程序的时候展示图形

```python
plt.plot(x,y)
plt.show
```

# 基础绘图和调整x轴刻度
 
  但是目前存在以下几个问题
- 1.设置图片大小(想要一个高清无码大图)
- 保存到本地
- 描述信息,比如x轴和y轴表示什么,这个图表示什么
- 调整x或者y的刻度的间距
- 线条的样式(比如颜色,透明度等)
- 标记出特殊的点(比如告诉别人最高点和最低点在哪里)
- 7.给图片添加一个水印(防伪,防止盗用)

- figure图形图标的意思,在这里指的就是我们画的图
- 通过实例化一个figure并且传递参数,能够在后台自动使用该figure实例
- 在图像模糊的时候可以传入dpi参数,让图片更加清晰

```python
# 设置图片大小
plt.figure(figsize=(20,8),dpi=80)

plt.plot(x,y)
#设置x轴的刻度.xticks(x)
plt.xticks(range(2,25))
#设置y轴的刻度.yticks(y)
plt.yticks(range(min(y),max(y)))
# 保存图片,有时可以保存为svg这种矢量格式,放大不会有锯齿
plt.savefig('./t1.png')
plt.show()
```

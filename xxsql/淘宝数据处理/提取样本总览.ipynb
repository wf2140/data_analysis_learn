```python
import os
import pandas as pd
class DataPerpare:
    def __init__(self,csv_path):
        self.csv_path = csv_path
    
    def create_sample(self,sample_size=1000000,output_path="taobao100w.csv"):
        print(f"创建样本数据({sample_size:,}行)")
        if not os.path.exists(self.csv_path):
            print(f"没找到文件 - {self.csv_path}")
            return False
        df_sample = pd.read_csv(
            self.csv_path,
            names = ['user_id', 'item_id', 'category_id', 'behavior_type', 'timestamp'],
            nrows = sample_size
        )
        df_sample.to_csv(output_path,index=False)

        print(f"已保存到{output_path}")
        print(f"样本大小{len(df_sample):,}行")
        print(f"时间范围:{pd.to_datetime(df_sample['timestamp'],unit='s').min()}到f{pd.to_datetime(df_sample['timestamp'],unit='s').max()}")
        return df_sample
    
    def analyze_data_base(self,csv_path):
        print("分析数据基本信息")
        df = pd.read_csv(csv_path)

        print(f"总行数:{len(df):,}")
        # print(f"占用内存:{df.memory_usage(deep=True).sum()/1024**2:.2f}MB")
        print("列信息:")
        for col in df.columns:
            null_count = df[col].isnull().sum()
            unique_count = df[col].nunique()
            print(f"{col}:")
            print(f"空值数:{null_count:,}({null_count/len(df)*100:.2f}%)")
            print(f"唯一值数:{unique_count:,}")
            if df[col].dtype != 'object':
                print(f"范围:{df[col].min()}到{df[col].max()}")
        return df
perpare = DataPerpare('data/UserBehavior.csv')
sample_df = perpare.create_sample()
df_analysis = perpareanalyze_data_base('taobao100w.csv')
```

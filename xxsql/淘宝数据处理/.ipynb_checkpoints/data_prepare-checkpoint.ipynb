```python
import os 
import zipfile
import pandas as pd
from tqdm import tqdm
# åˆ›å»ºæ•°æ®å‡†å¤‡è„šæœ¬ prepare_data.py

class DataPreparer:
    def __init__(self, zip_path, md5_path):
        self.zip_path = zip_path
        self.md5_path = md5_path
        self.csv_path = zip_path.replace('.zip', '')
        
    def verify_integrity(self):
        """éªŒè¯æ–‡ä»¶å®Œæ•´æ€§"""
        print("ğŸ” æ­£åœ¨éªŒè¯æ–‡ä»¶å®Œæ•´æ€§...")
        if not os.path.exists(self.md5_path):
            print("âš ï¸  è­¦å‘Šï¼šæœªæ‰¾åˆ°MD5æ ¡éªŒæ–‡ä»¶ï¼Œè·³è¿‡å®Œæ•´æ€§éªŒè¯")
            return True
            
        if verify_md5(self.zip_path, self.md5_path):
            return True
        else:
            print("\nå»ºè®®æ“ä½œï¼š")
            print("1. é‡æ–°ä¸‹è½½æ–‡ä»¶")
            print("2. æ£€æŸ¥ç½‘ç»œè¿æ¥")
            print("3. è”ç³»æ•°æ®æä¾›æ–¹")
            return False
    
    def extract_data(self):
        """è§£å‹æ•°æ®"""
        print(f"\nğŸ“¦ æ­£åœ¨è§£å‹ {self.zip_path}...")
        try:
            with zipfile.ZipFile(self.zip_path, 'r') as zip_ref:
                # æ˜¾ç¤ºå‹ç¼©åŒ…å†…å®¹
                print(f"å‹ç¼©åŒ…åŒ…å«ï¼š")
                for file_info in zip_ref.infolist():
                    print(f"  - {file_info.filename} ({file_info.file_size:,} bytes)")
                
                # è§£å‹
                zip_ref.extractall()
            print(f"âœ… è§£å‹å®Œæˆï¼")
            return True
        except zipfile.BadZipFile:
            print("âŒ å‹ç¼©æ–‡ä»¶æŸåï¼Œè¯·é‡æ–°ä¸‹è½½")
            return False
    
    def create_sample(self, sample_size=1000000, output_path="user_behavior_sample.csv"):
        """åˆ›å»ºæ ·æœ¬æ•°æ®"""
        print(f"\nğŸ¯ æ­£åœ¨åˆ›å»ºæ ·æœ¬æ•°æ®ï¼ˆ{sample_size:,} è¡Œï¼‰...")
        
        if not os.path.exists(self.csv_path):
            print("âŒ CSVæ–‡ä»¶ä¸å­˜åœ¨ï¼Œè¯·å…ˆè§£å‹")
            return False
        
        # è¯»å–æŒ‡å®šè¡Œæ•°çš„æ•°æ®
        df_sample = pd.read_csv(
            self.csv_path,
            names=['user_id', 'item_id', 'category_id', 'behavior_type', 'timestamp'],
            nrows=sample_size
        )
        
        # ä¿å­˜æ ·æœ¬
        df_sample.to_csv(output_path, index=False)
        
        print(f"âœ… æ ·æœ¬æ•°æ®å·²ä¿å­˜ï¼š{output_path}")
        print(f"   æ ·æœ¬å¤§å°ï¼š{len(df_sample):,} è¡Œ")
        print(f"   æ—¶é—´èŒƒå›´ï¼š{pd.to_datetime(df_sample['timestamp'], unit='s').min()} åˆ° "
              f"{pd.to_datetime(df_sample['timestamp'], unit='s').max()}")
        
        return df_sample
    
    def analyze_data_quality(self, csv_path):
        """åˆ†ææ•°æ®è´¨é‡"""
        print(f"\nğŸ“Š æ­£åœ¨åˆ†ææ•°æ®è´¨é‡...")
        
        # è¿™é‡Œå¯ä»¥æ·»åŠ å„ç§æ•°æ®è´¨é‡æ£€æŸ¥
        df = pd.read_csv(csv_path)
        
        print(f"æ€»è¡Œæ•°ï¼š{len(df):,}")
        print(f"å†…å­˜å ç”¨ï¼š{df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
        print("\nåˆ—ä¿¡æ¯ï¼š")
        for col in df.columns:
            null_count = df[col].isnull().sum()
            unique_count = df[col].nunique()
            print(f"  {col}:")
            print(f"    ç©ºå€¼æ•°ï¼š{null_count:,} ({null_count/len(df)*100:.2f}%)")
            print(f"    å”¯ä¸€å€¼æ•°ï¼š{unique_count:,}")
            if df[col].dtype != 'object':
                print(f"    èŒƒå›´ï¼š{df[col].min()} åˆ° {df[col].max()}")
        
        return df

# ä½¿ç”¨
preparer = DataPreparer("UserBehavior.csv.zip", "UserBehavior.csv.zip.md5")

# æ‰§è¡Œå®Œæ•´æµç¨‹
if preparer.verify_integrity():
    if preparer.extract_data():
        df = preparer.create_sample(500000)
        preparer.analyze_data_quality("user_behavior_sample.csv")
```

# Pandas对缺失值的处理

# Pandas使用这些函数处理缺失值:
 - isnull和notnull:检测是否是空值,可用于df和series
 - dropna:丢弃,删除缺失值
   - axis:删除行还是列,{0 or 'index',行 ; 1 or 'columns'列},default 0
   - how :如果等于any则任何值为空都删除,如果等于all则所有值都为空才删除
   - inplace:如果为True 则修改当前df,否则返回一个新的df
 - fillna:填充空值
   - vaue:用于填充的值,可以是单个值,或者字典(key是列名,value是值)
   - method:等于ffill使用前一个不为空的值填充forword fillL:等于bfill使用后一个不为空的值填充backword fill
   - axis:填充行还是列,{0 or 'index',行 ; 1 or 'columns'列},default 0
   - inplace:如果为True 则修改当前df,否则返回一个新的df

```python
import pandas as pd 
```

### 实例:特殊excel的读取,清洗和处理

#### 步骤一:读取excel的时候,忽略前几个空行

```python
studf = pd.read_excel("./数据清洗/成绩单.xlsx",skiprows = 2)
```

```python
studf
```

#### 步骤二:检测空值

```python
studf.isnull()
```

```python
studf['分数'].isnull()
```

```python
studf['分数'].notnull()
```

```python
#用来筛选没有空值的所有行
studf.loc[studf["分数"].notnull(),:]
```

#### 步骤三:删掉所有全是空值的列

```python
studf.dropna(axis = 1,how = 'all',inplace=True)
```

```python
studf
```

#### 步骤四:删掉全是空的行

```python
studf.dropna(axis = 0,how = 'all',inplace = True)
```

```python
studf
```

#### 步骤五:将分数列为空的填充为0分

```python
studf.fillna({"分数": 0})
```

```python
# 等同于:
studf.loc[:,'分数'] = studf['分数'].fillna(0)
studf
```

#### 步骤六:将姓名的缺失值填充:


 ##### 使用前面的有效值填充,用ffill:forward fill

```python
studf.loc[:,'姓名'] = studf['姓名'].fill(method = 'ffill')
现在用这个比较合适
studf.loc[:,'姓名'] = studf['姓名'].ffill()
studf
```

#### 步骤七:将清洗好的数据excel保存


```python
#index = False是为了不显示序号
studf.to_excel("数据清洗/成绩单1.xlsx",index = False)
```

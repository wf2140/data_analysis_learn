### 1.series

#### series是一种类似于一维数组的对象,它由一组数据以及一组与之相关的数据标签(及索引)组成

### 仅有数据列表即可产生的最简单的series

```python
import pandas as pd
import numpy as np
```

```python
s1 = pd.Series([1,'a',33.4,8])
#左侧是索引,右侧是数据
s1
```

```python
#获取索引
s1.index
```

```python
#获取数据
s1.values
```

### 创建一个具有标签索引的Series

```python
a2 = pd.Series([1,'a',33.2,7],index=['d','b','c','a'])
```

```python
a2
```

#### 使用python字典创建Series

```python
sdata={'xw':3500,'xz':4000,'xt':5000,'xa':7000}
```

```python
a3 = pd.Series(sdata)
a3
```

### 根据标签索引查询数据
#### 类似python的字典

```python
a2
```

```python
a2['a']
```

```python
type(a2['a'])
```

```python
a2[['a','b']]
```

```python
type(a2[['a','b']])
```

## 2. DataFrame
### DataFrame是一个表格型的数据结构
#### 。 每列可以是不同的值类型 数值、字符串、布尔值等)
#### 。 既有行索引index,也有列索引columns
#### 。可以被看做由Series组成的字典
### 创建dataframe最常用的方法，见02节读取纯文本文件、excel、mysql数据库
### 2.1 根据多个字典序列创建dataframe

```python
data = {
    'state':['a','a','a','b','b'],
    'year':[2000,2001,2002,2001,2002],
    'pop':[1.5,1.7,3.6,2.4,2.9]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.dtypes
```

```python
df.columns
```

```python
df.index
```
